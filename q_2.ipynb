Random numbers using np.random.rand()

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# Record start time
start_time = time.time()

# Generate 10,000 random numbers between 0 and 1
random_numbers = np.random.rand(10000)

# Record end time
end_time = time.time()

# Calculate elapsed time
elapsed_time = end_time - start_time

print("Time taken to generate 10,000 uniform deviates using np.random.rand():", elapsed_time, "seconds")

# Create a histogram
plt.hist(random_numbers, range=(0.0, 1.0), density = True)

#Generate the uniform PDF
x = np.linspace(0, 1, 100)
plt.plot(x, np.ones_like(x), color='red', linestyle='--', label = 'uniform PDF')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Random Numbers (numpy)')
plt.legend()

# Show plot
plt.show() 
```
