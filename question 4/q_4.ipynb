```python
import numpy as np
import matplotlib.pyplot as plt

# Load the generated samples
samples = np.loadtxt('exponential_samples.txt')

# Plot the histogram of the samples
plt.hist(samples, bins=50, density=True, alpha=0.6, color='g', label='Sampled Data')

# Plot the exponential PDF
mean = 0.5
lam = 1 / mean
x = np.linspace(0, max(samples), 1000)
pdf = lam * np.exp(-lam * x)
plt.plot(x, pdf, 'r-', lw=2, label='Exponential PDF')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Exponential Distribution (mean=0.5)')
plt.legend()
plt.show()
```
