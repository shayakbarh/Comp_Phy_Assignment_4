Box-Muller method to produce 10,000 random numbers distributed according to a Gaussian distribution

```python
import numpy as np
import matplotlib.pyplot as plt

x1 = np.random.rand(10000)
x2 = np.random.rand(10000)

y1 = np.sqrt(- 2 * np.log(x1)) * np.cos( 2 * np.pi * x2)
y2 = np.sqrt(- 2 * np.log(x1)) * np.sin( 2 * np.pi * x2)

# Generate the Gaussian PDF
x = np.linspace(-5, 5, 100)
pdf = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*x**2)


# Set the figure size
plt.figure(figsize=(10, 6))

# Plot histogram for y1
plt.subplot(1,2,1)
plt.hist(y1, density = True, color = 'c')
plt.plot(x, pdf, color='red', linestyle='--', label = 'Gaussian PDF')
plt.xlabel("y1")
plt.ylabel("PDF")
plt.title("Density Histogram of 10000 Normal Random Numbers")
plt.legend()

# Plot histogram for y2
plt.subplot(1,2,2)
plt.hist(y2, density = True, color = 'b')
plt.plot(x, pdf, color='red', linestyle='--', label = 'Gaussian PDF')
plt.xlabel("y2")
plt.ylabel("PDF")
plt.title("Density Histogram of 10000 Normal Random Numbers")
plt.legend()

plt.tight_layout()
plt.show()
```
