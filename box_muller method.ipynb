```python
import numpy as np
import matplotlib.pyplot as plt

# Number of random numbers to generate
n = 5000

# Generate uniform random numbers in [0, 1]
u1 = np.random.rand(n)
u2 = np.random.rand(n)

# Box-Muller transformation to convert uniform random numbers to standard normal random numbers
z1 = np.sqrt(-2 * np.log(u1)) * np.cos(2 * np.pi * u2)
z2 = np.sqrt(-2 * np.log(u1)) * np.sin(2 * np.pi * u2)

# Create an array of normal random numbers with mean 0 and variance 1 by combining z1 and z2
normal_numbers = np.concatenate((z1, z2))

# Create a histogram
plt.hist(normal_numbers, density=True, alpha=0.7, color='blue', edgecolor='black')

# Generate the Gaussian PDF
x = np.linspace(-5, 5, 100)
pdf = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*x**2)

# Plot the Gaussian PDF
plt.plot(x, pdf, color='red', linestyle='--')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Normal Random Numbers')

# Show plot
plt.show()

print(len(normal_numbers))
```
