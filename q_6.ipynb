```python
import numpy as np
import matplotlib.pyplot as plt

# given distribution
def f(x):
    return np.sqrt(2/np.pi) * np.exp(-x**2)

# enveloped function
def g(x):
    if 0 < x < 10:
        return 1
    else:
        return 0


xarr = np.linspace(0,10,10000)

farr = [f(i) for i in xarr]
garr = [ g(i) for i in xarr]

# Plot f(x) and g(x)
plt.plot(xarr, farr, label = 'Distribution function f(x)', color = 'b')
plt.plot(xarr, garr, color = 'k', label = 'Enveloped function g(x)')


# Generate random numbers
x = np.random.rand(10000) * 10
y = np.random.rand(10000) * garr

# accepted points
y_good = y[ y < f(x)]
x_good = x[ y < f(x)]

# rejected points
y_bad = y[ y > f(x)]
x_bad = x[ y > f(x)]

#plt.scatter(x, y, color = 'm', marker = '.')
plt.scatter(x_good, y_good, color = 'g', marker = '.', label = 'accepted points')
#plt.scatter(x_bad, y_bad, color = 'r', marker = '.', label = 'rejected points')

plt.hist(x_good, range = (0.0, 10.0), density = True, label = 'density histogram', color = 'orange')
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Random numbers using Rejection method")
plt.legend()
plt.show()
```
