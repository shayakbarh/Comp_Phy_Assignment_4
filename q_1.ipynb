Linear congruential random number generator

```python
import numpy as np
import matplotlib.pyplot as plt
import time

# linear congruential generator to generate numbers between 0 and 1
def linear_congruential_generator(x0, n, a = 1103515245, c = 12345, m = 2**31):  # values of a, c, m are same with rand() of C
    results = []
    x = x0

    for i in range(n):
        x = (a * x + c) % m
        results.append(x)

    random_numbers = [number / (m-1) for number in results]

    return random_numbers    

# record start time
start_time = time.time()

# generate random numbers using LC generator
random_numbers = linear_congruential_generator(0.5, 10000)

# record end time
end_time = time.time()

# calculate time taken
elapsed_time = end_time - start_time

print("Time taken to generate 10,000 uniform deviates using LC generator: ", elapsed_time, "seconds")


# Create a histogram
plt.hist(random_numbers, range=(0.0, 1.0), density = True)

#Generate the uniform PDF
x = np.linspace(0, 1, 100)
plt.plot(x, np.ones_like(x), color='red', linestyle='--', label = 'uniform PDF')

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of Random Numbers (LC Generator)')
plt.legend()
# Show plot
plt.show()   
```
